# 01 — EDA: Tempo de Titulação

Análise exploratória da distribuição do tempo de conclusão dos discentes de Sistemas de Informação da UFRN (2009–2026).

```python
import sys
sys.path.append('..')  # garante que src/ seja encontrado

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data.extrair import carregar_discentes_si
from src.features.build_features import construir_features
from src.config import CSV_DISCENTES_SI
```

```python
# Carrega e processa os dados
df_bruto = pd.read_csv(CSV_DISCENTES_SI)
df = construir_features(df_bruto)
print(f'Total de alunos: {len(df)}')
df.head()
```

```python
# Histograma: concluintes por semestres cursados
df_conc = df[df['flag_concluinte'] == 1]
hist = df_conc[df_conc['semestres_cursados'].between(8, 20)]

fig, ax = plt.subplots(figsize=(10, 5))
hist['semestres_cursados'].value_counts().sort_index().plot(kind='bar', ax=ax, color='#2196F3')
ax.set_title('Concluintes por Semestres Cursados')
ax.set_xlabel('Semestres')
ax.set_ylabel('Quantidade')
plt.tight_layout()
plt.show()
```
